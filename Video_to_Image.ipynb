```
def length_of_video(video_path):
  cap = cv2.VideoCapture(video_path)
  length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  return length
```

```
def extracting_frames(video_path, save_path, skip_frames = 0):
  length = length_of_video(video_path)
  if length == 0:
    print("Your video has no length, cannot extract frames")
    return 0

  cap = cv2.VideoCapture(video_path)

  ret, frame = cap.read()

  count = 0
  while ret:
    ret, frame = cap.read()
    if ret:
      cv2.imwrite(my_drive + "/Photos/"+ str(count)+ ".jpg", frame)
      count += 1
      print(str(round((round(count/(length),3)*100))) + "% Complete")
    else:
      count+=1

  cap.release()
```

```
def frame_ext(form, name):
  if form == False:
    video = my_drive + str(name)
    print(video)
    print(length_of_video(video))
    extracting_frames(video, my_drive, skip_frames = 0)
  elif form == True:
    video = form_loc + str(name)
    print(video)
    print(length_of_video(video))
    extracting_frames(video, my_drive, skip_frames = 0)
```
