```python
from google.colab import drive
drive.mount('/content/gdrive')
my_drive = '/content/gdrive/My Drive/Colab Notebooks/'
```

```python
def video_assembly(save_path, count, output="/Image Colorization/Video Input/",fps=30):
  img_array = []
  num = 0
  while num < count:
    img = cv2.imread(my_drive + output + "Photos/"+ str(num) + ".jpg")
    height, width, layers = img.shape
    size = (width, height)
    img_array.append(img)
    num += 1
    print(str(round((round(num/(count),3)*100))) + "% Complete")

  out = cv2.VideoWriter(my_drive + output +'Colorized_Video.avi',cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
 
  for i in range(len(img_array)):
    out.write(img_array[i])
  out.release()

  return out
```

```python
video_assembly(my_drive, 224, output="/Image Colorization/Video Output/")
```
